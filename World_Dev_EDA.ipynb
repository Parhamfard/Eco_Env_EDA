2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

    Write a function that takes in values and returns the standard deviation of the values using the formula above. Hint: use the function you wrote above to calculate the mean! Use the list of numbers you created above to test on your function.
    
    *Note*: Do not use any standard deviation methods built-in to any Python libraries to do this! This should be done without importing any additional libraries.

```python
# Code standard deviation: Calculate the standard deviation of each column in the data frame and using mean function
def std_dev(data):
    means = mean(data)
    std_devs = {}
    for column in data:
        column_data = data[column]
        mean_value = means[column]
        variance = sum((x - mean_value) ** 2 for x in column_data) / len(column_data)
        std_devs[column] = variance ** 0.5  
    return std_devs    
```

3. Data cleaning function:
    
    Write a function that takes in a string consisting a number and an abbreviated place value (ex. '5M', '45K'), and converts that string to an integer. For example, inputting '5M' in your function should return 5000000, '45K' should return 45000, etc. Make sure to test your function to make sure it works!

You will use these functions later on in the project!

```python
#Code Function takes string and returns float
def convert_to_float(value):
    try:
        if isinstance(value, str):
            value = value.strip()  
            if value.endswith('M') or value.endswith('m'):
                return float(value[:-1]) * 1000000
            elif value.endswith('K') or value.endswith('k'):
                return float(value[:-1]) * 1000
        return float(value)  
    except (ValueError, TypeError):
        return None  
```

```python
# Function to calculate missing values count, percentage, and sum
def missing_values_info(df, label):
    missing_count = df.isnull().sum()
    missing_percentage = (missing_count / len(df)) * 100
    missing_sum = missing_count.sum()  # Sum of missing values across the entire dataframe
    return pd.DataFrame({
        f"{label}_missing_count": missing_count,
        f"{label}_missing_percentage": missing_percentage,
        f"{label}_missing_sum": missing_sum  # Add the sum of missing values
    })

# Calculate and merge missing values info for each dataframe
dataframes = [
    (df_converted_population, "data_population"),
    (df_converted_life_expec, "data_life_expec"),
    (df_converted_gni_cap, "data_gni_per_cap"),
    (df_converted_co2_cap, "data_co2_per_cap")
]

missing_combined = pd.concat(
    [missing_values_info(df, label) for df, label in dataframes],
    axis=1,
    join='outer'  # Ensure all columns are aligned correctly
)

# Print the combined missing values information
print(missing_combined)

# Sum of missing values for each dataframe
missing_sums = {label: df.isnull().sum().sum() for df, label in dataframes}

# Print sum of missing values for each dataframe
print("Sum of missing values per dataframe:")
for label, total_missing in missing_sums.items():
    print(f"{label}: {total_missing}")
```

### Testing Different Imputation Methods for China and India Population Data

I tested four methods for imputing missing values in the population data for both China and India:

1. **Linear Interpolation**
2. **Polynomial Interpolation**
3. **Spline Interpolation**
4. **Extrapolation (Mean)**


```python
# Extract China data (rows where country is 'China')
china_data = df_converted_population[df_converted_population['country'] == 'China'].iloc[:, 1:-1].values.flatten()

# Extract India data (rows where country is 'India')
india_data = df_converted_population[df_converted_population['country'] == 'India'].iloc[:, 1:-1].values.flatten()

# Extract years (column names from 1800 to 2100)
years = df_converted_population.columns[1:-1].astype(int)
```

```python
# Convert data to a pandas Series for interpolation
china_series = pd.Series(china_data)
india_series = pd.Series(india_data)

# Apply linear interpolation to fill missing data
china_linear_interp = china_series.interpolate(method='linear')
india_linear_interp = india_series.interpolate(method='linear')

# Print results to verify
print("China Linear Interpolation:", china_linear_interp.head())
print("India Linear Interpolation:", india_linear_interp.head())
```

```python
# Apply polynomial interpolation (degree 3)
china_poly_interp = china_series.interpolate(method='polynomial', order=3)
india_poly_interp = india_series.interpolate(method='polynomial', order=3)

# Print results to verify
print("China Polynomial Interpolation:", china_poly_interp.head())
print("India Polynomial Interpolation:", india_poly_interp.head())
```

```python
# Apply spline interpolation (degree 3)
china_spline_interp = china_series.interpolate(method='spline', order=3)
india_spline_interp = india_series.interpolate(method='spline', order=3)

# Print results to verify
print("China Spline Interpolation:", china_spline_interp.head())
print("India Spline Interpolation:", india_spline_interp.head())
```

```python
# Apply spline interpolation (degree 3)
china_spline_interp = china_series.interpolate(method='spline', order=3)
india_spline_interp = india_series.interpolate(method='spline', order=3)

# Print results to verify
print("China Spline Interpolation:", china_spline_interp.head())
print("India Spline Interpolation:", india_spline_interp.head())
```

### Comparison with Methods from the Gapminder Paper

In the Gapminder paper titled *"Documentation for Total Population for Countries and Territories"* (PDF file: **gapdoc003.pdf**, pages 10-13), the following methods are discussed for filling missing data:

1. **Interpolation and Extrapolation Techniques**:
   - The document suggests using **geographical and temporal interpolation or extrapolation** to estimate missing population values. Geographical interpolation involves using the growth trends of neighboring regions or countries, while temporal interpolation connects two known data points to estimate missing values. Temporal extrapolation extends trends from the available data either forward or backward.
   
2. **Arbitrary Guesses**:
   - When no reliable estimation methods are available, the paper proposes using **arbitrary guesses**. This approach sets the population for a missing observation equal to the earliest available observation, ensuring consistency in graphical displays and showing population trends over time.

However, in my analysis of the population data for **China** and **India**, I found that **interpolation and extrapolation techniques** did not yield satisfactory results for both countries. Additionally, the use of **arbitrary guesses** was not a precise approach and did not provide reliable predictions.

```python
# Function to fill missing data for China and India using their census data
def fill_missing_population_data(df, china_census, india_census):
    # Remove leading and trailing spaces from column names in census dataframes
    china_census.columns = china_census.columns.str.strip()
    india_census.columns = india_census.columns.str.strip()
    
    # Iterate through the rows of df_converted_population_cleaned
    for index, row in df.iterrows():
        country = row['country']  # Get the country name
        
        # If the country is China, fill missing data with the China census data
        if country == 'China':
            for year in range(1950, 2101):  # For years 1950 to 2100
                if pd.isnull(row[str(year)]):  # If data is missing for the year
                    df.at[index, str(year)] = china_census.loc[china_census['Years'] == year, 'China'].values[0]

        # If the country is India, fill missing data with the India census data
        elif country == 'India':
            for year in range(1950, 2101):  # For years 1950 to 2100
                if pd.isnull(row[str(year)]):  # If data is missing for the year
                    df.at[index, str(year)] = india_census.loc[india_census['Years'] == year, 'India'].values[0]

    return df

# Apply the function to fill missing data
df_converted_population_final = fill_missing_population_data(df_converted_population, China_population_census, India_population_census)

# Now that we have the final population data, let's visualize the population of China and India

# Extract the population data for China and India for years 1950 to 2100
china_population = df_converted_population_final[df_converted_population_final['country'] == 'China'].iloc[:, 1:].transpose()
india_population = df_converted_population_final[df_converted_population_final['country'] == 'India'].iloc[:, 1:].transpose()

# Rename the columns to make them clearer
china_population.columns = ['China']
india_population.columns = ['India']

# Combine both country populations into a single DataFrame
population_data = pd.concat([china_population, india_population], axis=1)

# Create the plot
plt.figure(figsize=(14, 7))  # Increase the figure size to make the labels more readable
plt.plot(population_data.index, population_data['China'], label='China', color='red', linewidth=2)
plt.plot(population_data.index, population_data['India'], label='India', color='blue', linewidth=2)

# Adding titles and labels
plt.title('Population of China and India (1950-2100)', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Population', fontsize=12)
plt.legend()

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Save the plot as a PNG image
plt.savefig('population_of_china_and_india_1950_2100.png', dpi=300, bbox_inches='tight')

# Show the plot
plt.grid(True)
plt.tight_layout()  # Adjust the layout to avoid clipping
plt.show()
```

- A correlation close to **0** indicates **no relationship** or a **weak correlation** between the variables.

A **scatter plot** is a graph that shows individual data points as dots, where each dot represents a country. The position of the dot is determined by two indicators:

The analysis reveals interesting relationships between the four key indicators in G20 countries:

- **Life Expectancy and Population**: The correlation was weak and negative (-0.35 in 1970, -0.23 in 2020), suggesting that larger populations do not necessarily correlate with better life expectancy.
- **CO2 Emissions per Capita and Population**: A weak negative correlation was observed, highlighting that larger populations are not always associated with higher emissions.
  
---

### 2. **Changes in Correlation Patterns from 1970 to 2020**

The correlations have evolved over the years:

- **Life Expectancy and Economic Indicators**: While economic growth once strongly contributed to better life expectancy, the relationship has weakened by 2020, likely due to advancements in healthcare and technology that decouple life expectancy from purely economic growth.
- **CO2 Emissions and Economic Growth**: In 1970, economic growth and CO2 emissions were strongly linked, but by 2020, the relationship weakened as cleaner technologies and stronger environmental policies allowed countries to grow economically with reduced emissions.
- **Population Size and Life Expectancy**: The relationship between population size and life expectancy has remained weak, but large populations no longer pose as significant a challenge to life expectancy improvements, due to advancements in healthcare and urban planning.
